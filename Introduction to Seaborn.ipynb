```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
stats = pd.read_csv("Demographic.csv")
stats.head()
plt.rcParams['figure.figsize'] = 8,4
```

```python
stats.columns
```

```python
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',
       'IncomeGroup']
```

```python
vis1 = sns.distplot(stats["InternetUsers"]) #histogram
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=30)
```

```python
# for boxplot
sns.set_style("darkgrid")
vis2 = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate")
```

```python
vis3 = sns.lmplot(x="InternetUsers", y="BirthRate", data=stats, fit_reg=False, hue="IncomeGroup", size=4, aspect=2)
```

```python
sns.swarmplot(data=stats, x="IncomeGroup", y="BirthRate") # for categorical data
```
